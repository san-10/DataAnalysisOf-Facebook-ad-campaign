Handling Missing Data

```
fbdata_df.isnull().sum()
```

```
fbdata_df["campaign_id"] = fbdata_df.groupby('gender')['campaign_id'].transform(lambda x: x.ffill())  
```

- To know the missing data and filling the rows with ffill meothed.

```
fbdata_df
```

int /float

```
fbdata_df['campaign_id']=fbdata_df['campaign_id'].fillna(0).astype(int)
fbdata_df.tail()
```

```
fbdata_df['clicks']=fbdata_df['clicks'].fillna(0).astype(int)
fbdata_df.tail()
```

- Converting  values in campaign_id column from float to int.
